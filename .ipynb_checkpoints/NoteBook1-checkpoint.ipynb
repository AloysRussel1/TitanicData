```python
data = data.rename(columns={'Survived': 'Survival_Status'})
```

```python
print("Colonnes :", data.columns)
```

```python
femmes = data[data['Sex'] == 'female']
```

```python
enfants = data[data['Age'] < 18]
```

```python
hommes_adultes = data[(data['Sex'] == 'male') & (data['Age'] >= 18)]
```

```python
taux_survie_femmes = femmes['Survival_Status'].mean()
taux_survie_enfants = enfants['Survival_Status'].mean()
taux_survie_hommes_adultes = hommes_adultes['Survival_Status'].mean()
```

```python
print(f"Taux de survie des femmes : {taux_survie_femmes:.2%}")
print(f"Taux de survie des enfants : {taux_survie_enfants:.2%}")
print(f"Taux de survie des hommes adultes : {taux_survie_hommes_adultes:.2%}")
```

```python
if taux_survie_femmes > taux_survie_hommes_adultes and taux_survie_enfants > taux_survie_hommes_adultes:
    print("La priorité semble avoir été donnée aux femmes et aux enfants.")
else:
    print("Aucune preuve claire de priorité pour les femmes et les enfants.")
```
